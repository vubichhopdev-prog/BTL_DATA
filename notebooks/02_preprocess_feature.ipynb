```python
import pandas as pd
import os

print("-> BẮT ĐẦU QUY TRÌNH TIỀN XỬ LÝ DỮ LIỆU...")

# 1. Đọc dữ liệu thô
df_raw = pd.read_csv('../data/raw/ai4i2020.csv')

# 2. Loại bỏ các cột định danh không mang ý nghĩa huấn luyện mô hình
cols_to_drop = ['UDI', 'Product ID']
df_clean = df_raw.drop(columns=cols_to_drop)
print("- Đã xóa các cột định danh.")

# 3. Mã hóa (Encoding) biến phân loại 'Type'
# Chất lượng L (Low) = 0, M (Medium) = 1, H (High) = 2
df_clean['Type'] = df_clean['Type'].map({'L': 0, 'M': 1, 'H': 2})
print("- Đã mã hóa cột Type (L/M/H -> 0/1/2).")

# 4. Kiểm tra thư mục lưu trữ
out_dir = '../data/processed'
if not os.path.exists(out_dir):
    os.makedirs(out_dir)

# 5. Lưu dữ liệu đã làm sạch
out_path = f'{out_dir}/ai4i2020_processed.csv'
df_clean.to_csv(out_path, index=False)
print(f"-> HOÀN TẤT! Đã lưu dữ liệu sạch vào: {out_path}")

# Hiển thị thử 5 dòng đầu
display(df_clean.head())
```
