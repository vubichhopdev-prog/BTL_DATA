```python
import sys
import os
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sys.path.append(os.path.abspath('..'))

from src.data.loader import DataLoader
from src.models.forecasting import ToolWearPredictor

sns.set_theme(style="whitegrid")
```

```python
# Đọc dữ liệu thô
DATA_PATH = '../data/raw/ai4i2020.csv'
loader = DataLoader(DATA_PATH)
df = loader.load_data()

# Khởi tạo mô hình dự đoán mòn dao
predictor = ToolWearPredictor()

# Chuẩn bị dữ liệu (Bắt buộc không shuffle theo rubric)
X_train, X_test, y_train, y_test = predictor.prepare_data(df)
print(f"Số lượng Train: {len(X_train)} dòng | Số lượng Test: {len(X_test)} dòng")

# Huấn luyện mô hình dự đoán
predictor.train(X_train, y_train)

# Đánh giá kết quả và xuất hình ảnh biểu đồ
predictor.evaluate(X_test, y_test)
```
